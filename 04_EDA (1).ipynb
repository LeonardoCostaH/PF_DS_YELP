### California 1/4

```python
# Import
scrap_california_hotels = pd.read_csv("Datos\\Scraping\\raw_California_hotels.csv")
```

```python
df_scrap_california = pd.DataFrame(scrap_california_hotels)
```

```python
df_scrap_california.head()
```

```python
# Avg_score
df_scrap_california['avg_score'] = df_scrap_california['avg_score'].str.replace(',', '.')
# Remove non-numeric characters from 'avg_score' column
df_scrap_california['avg_score'] = df_scrap_california['avg_score'].replace('[^\d.]', '', regex=True)
# Convert 'avg_score' to float
df_scrap_california['avg_score'] = df_scrap_california['avg_score'].astype(float)

#Price
df_scrap_california['price'] = df_scrap_california['price'].str.replace('$', '')
df_scrap_california['price'] = df_scrap_california['price'].str.replace('.', '')
# Convert 'price' column to int
df_scrap_california['price'] = pd.to_numeric(df_scrap_california['price'], errors='coerce')
```

```python
# Cambiando el nombre de url por reviews_url
df_scrap_california.rename(columns={'url': 'reviews_url'}, inplace=True)
```

### Utah 4/4

```python
df_scrap_utah = pd.read_csv("data/booking/raw_Utah_hotels.csv")

# Dropping rows where 'name' is NaN
df_scrap_utah = df_scrap_utah.dropna(subset=['name'])

# Removing duplicates from 'url' column
df_scrap_utah['reviews_url'] = df_scrap_utah['reviews_url'].drop_duplicates()

# Avg_score
df_scrap_utah['avg_score'] = df_scrap_utah['avg_score'].replace(',', '.')
# Remove non-numeric characters from 'avg_score' column
df_scrap_utah['avg_score'] = df_scrap_utah['avg_score'].replace('[^\d.]', '', regex=True)
# Convert 'avg_score' to float
df_scrap_utah['avg_score'] = df_scrap_utah['avg_score'].astype(float)
```

```python
#Price
for i, row in df_scrap_utah.iterrows():

    price = row['price']
    price = price.replace('$', '')
    price = price.replace(' ', '')

    if price.count(".") == 1:
        df_scrap_utah.at[i, 'price'] = price
    elif price.count(".") == 2:
        first_dot_index = price.find('.')
        df_scrap_utah.at[i, 'price'] = price[:first_dot_index] + price[first_dot_index+1:]
    else:
        print(price)

# Convert 'price' column to int
df_scrap_utah['price'] = pd.to_numeric(df_scrap_utah['price'], errors='coerce')
```

# Atracciones turisticas Utah

```python
scrap_utah_atracciones = pd.read_csv("Datos\\Scraping\\Utah_attractions.csv")
```

```python
df_utah_attractions = pd.DataFrame(scrap_utah_atracciones)
```

```python
df_utah_attractions.head()
```

```python
# Remove leading numbers and dots
df_utah_attractions['attraction'] = df_utah_attractions['attraction'].str.replace(r'^\d+\.\s*', '', regex=True)
```

```python
import unidecode
# Replace '•' with commas
df_utah_attractions['categories'] = df_utah_attractions['categories'].str.replace(' • ', ', ')
# Remove diacritic marks
df_utah_attractions['categories_list'] = df_utah_attractions['categories_list'].apply(unidecode.unidecode)

# Create a new column with lists of categories
df_utah_attractions['categories_list'] = df_utah_attractions['categories'].str.split(', ')
```

```python
df_utah_attractions['categories_list']
```

```python
# Concatenate all lists of attractions into a single list
all_attractions = [attraction for attractions_list in df_utah_attractions['categories_list'] for attraction in attractions_list]

# Remove duplicates by converting the list to a set and then back to a list
unique_attractions = list(set(all_attractions))

# Display the list of unique attractions
unique_attractions
```

```python
from collections import Counter

# Concatenate all lists of attractions into a single list
all_attractions = [attraction for attractions_list in df_utah_attractions['categories_list'] for attraction in attractions_list]

# Use Counter to count occurrences of each category
attraction_counts = Counter(all_attractions)

# Display the counts
attraction_counts
```

```python
import matplotlib.pyplot as plt

# Concatenate all lists of attractions into a single list
all_attractions = [attraction for attractions_list in df_utah_attractions['categories_list'] for attraction in attractions_list]

# Use Counter to count occurrences of each category
attraction_counts = Counter(all_attractions)

# Convert the counts dictionary to two lists for plotting
categories, counts = zip(*attraction_counts.items())

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(categories, counts, color='skyblue')
plt.xlabel('Attraction Category')
plt.ylabel('Count')
plt.title('Attraction Categories Count')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.tight_layout()

# Show the plot
plt.show()
```
