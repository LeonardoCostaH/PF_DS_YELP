```python
def cargar_json(ruta):
    with open(ruta, 'r', encoding='utf-8') as archivo:
        contenido = archivo.read()
        # Utilizar json.loads para cargar cada objeto JSON por separado en una lista
        contenido_json = [json.loads(line) for line in contenido.splitlines()]

    # Convertir la lista de objetos JSON en un DataFrame de pandas
    dataframe = pd.json_normalize(contenido_json)

    return dataframe
```
