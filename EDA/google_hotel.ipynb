## CATEGORIES

```python
data_google['categories'] = data_google['categories'].str.lower()
```

```python
def convertir_a_lista(cadena):
    return cadena.split(',')

# Aplicar la función de conversión a la columna 'categorias'
data_google['categories'] = data_google['categories'].apply(convertir_a_lista)
```

```python
df_exploded = data_google.explode(column="categories")
```

```python
df_exploded.info()
```

```python
df_exploded.tail(20)
```

```python
df_exploded.isnull().sum()
```

```python
df_exploded["categories"].unique().tolist()
```

```python
data_filt= df_exploded.copy()
```

```python
palabras_clave = ['hotel','motel', 'lodge', 'resort', 'hostel', 'bed and breakfast', 'guesthouse','retreat','boutique']

# Filtrar las filas basándote en las palabras clave en la columna 'categorias'
filtro = data_filt['categories'].str.contains('|'.join(palabras_clave), case=False)

# Aplicar el filtro al DataFrame
df_filtrado = data_filt[filtro]
```

```python
df_filtrado
```

```python
df_filtrado = df_filtrado.drop_duplicates(subset=['business_id'])
```

```python
df_filtrado
```

```python
df_filtrado["categories"].unique().tolist()
```

```python
df_filtrado = df_filtrado.apply(lambda x: x.str.replace("[", '') if x.dtype == "object" else x)
df_filtrado = df_filtrado.apply(lambda x: x.str.replace("]", '') if x.dtype == "object" else x)
df_filtrado = df_filtrado.apply(lambda x: x.str.replace("'", '') if x.dtype == "object" else x)
```

```python
df_filtrado
```

```python
df_filtrado=df_filtrado.drop("Unnamed: 0", axis=1)
```

```python
df_filtrado.to_csv("google_hotel.csv", index=False)
```
