```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

### California 1/4

```python
# Import
scrap_california_hotels = pd.read_csv("Datos\\Scraping\\raw_California_hotels.csv")
```

```python
df_scrap_california = pd.DataFrame(scrap_california_hotels)
```

```python
df_scrap_california.head()
```

```python
df_scrap_california.info()
```

```python
# Avg_score
df_scrap_california['avg_score'] = df_scrap_california['avg_score'].str.replace(',', '.')
# Remove non-numeric characters from 'avg_score' column
df_scrap_california['avg_score'] = df_scrap_california['avg_score'].replace('[^\d.]', '', regex=True)
# Convert 'avg_score' to float
df_scrap_california['avg_score'] = df_scrap_california['avg_score'].astype(float)

#Price
df_scrap_california['price'] = df_scrap_california['price'].str.replace('$', '')
df_scrap_california['price'] = df_scrap_california['price'].str.replace('.', '')
# Convert 'price' column to int
df_scrap_california['price'] = pd.to_numeric(df_scrap_california['price'], errors='coerce')
```

```python
scrap_california_hotels.shape
```

### Utah 4/4

```python
scrap_utah_hotels = pd.read_csv("Datos\\Scraping\\raw_Utah_hotels.csv")
```

```python
df_scrap_utah = pd.DataFrame(scrap_utah_hotels)
```

```python
df_scrap_utah.info()
```

```python
df_scrap_utah.head()
```

```python
# Dropping rows where 'name' is NaN
df_scrap_utah = df_scrap_utah.dropna(subset=['name'])
```

```python
# Avg_score
df_scrap_utah['avg_score'] = df_scrap_utah['avg_score'].str.replace(',', '.')
# Remove non-numeric characters from 'avg_score' column
df_scrap_utah['avg_score'] = df_scrap_utah['avg_score'].replace('[^\d.]', '', regex=True)
# Convert 'avg_score' to float
df_scrap_utah['avg_score'] = df_scrap_utah['avg_score'].astype(float)

#Price
df_scrap_utah['price'] = df_scrap_utah['price'].str.replace('$', '')
df_scrap_utah['price'] = df_scrap_utah['price'].str.replace('.', '')
# Convert 'price' column to int
df_scrap_utah['price'] = pd.to_numeric(df_scrap_utah['price'], errors='coerce')
# Removing duplicates from 'url' column
df_scrap_utah['url'] = df_scrap_utah['url'].drop_duplicates()
```

```python
# Removing duplicates from 'url' column
df_scrap_utah['url'] = df_scrap_utah['url'].drop_duplicates()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Creating a scatter plot with regression line
sns.regplot(x='price', y='avg_score', data=df_scrap_utah,color='green')
```

```python
import seaborn as sns
sns.set_theme(style="darkgrid")

g = sns.jointplot(x="price", y="avg_score", data=df_scrap_utah,
                  kind="reg", truncate=False, color='purple')
```

Hacer un mapa de distribucion de precios y avg_score de acuerdo a la ciudad
histograma de scores mas comunes

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df_scrap_utah is your DataFrame

# Set the figure size
plt.figure(figsize=(10, 6))

# Create a scatter plot
sns.scatterplot(x='price', y='avg_score', hue='city', data=df_scrap_utah, palette='viridis', alpha=0.7)

# Set labels and title
plt.xlabel('Price')
plt.ylabel('Average Score')
plt.title('Scatter Plot of Price vs. Average Score by City')

# Show the legend
plt.legend(title='City', loc='upper right')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the figure size
plt.figure(figsize=(12, 6))

# Create a boxplot for Average Score by City
sns.boxplot(x='city', y='avg_score', data=df_scrap_utah, palette='viridis')

# Set labels and title
plt.xlabel('City')
plt.ylabel('Average Score')
plt.title('Boxplot of Average Score by City')

# Show the plot
plt.show()
```

```python
# Set the figure size
plt.figure(figsize=(12, 6))

# Create a boxplot for Price by City
sns.boxplot(x='city', y='price', data=df_scrap_utah, palette='viridis')

# Set labels and title
plt.xlabel('City')
plt.ylabel('Price')
plt.title('Boxplot of Price by City')

# Show the plot
plt.show()
```
