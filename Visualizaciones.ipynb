```python
# import data
yelp_hotel_business_viz = pd.read_csv("Datos\\Yelp\\drive\\yelp_hotel_business.csv")
```

```python
# Creando df
df_yelp_hotel_business_viz = pd.DataFrame(yelp_hotel_business_viz)
```

```python
df_yelp_hotel_business_viz.head()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame

df = df_yelp_hotel_business_viz

# Convert 'avg_rating' to numeric (if not already)
df['avg_rating'] = pd.to_numeric(df['avg_rating'], errors='coerce')

# Calculate the average of 'avg_rating' by 'state'
avg_ratings_by_state = df.groupby('state')['avg_rating'].mean().reset_index()

# Set Seaborn theme
sns.set_theme(style="whitegrid")

# Bar plot of average ratings by state
plt.figure(figsize=(10, 5))
sns.barplot(x='state', y='avg_rating', data=avg_ratings_by_state, ci=None, color='purple')
plt.title('Average Ratings by state (Bar Plot)')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = df_yelp_hotel_business_viz

# Convert 'avg_rating' to numeric (if not already)
df['avg_rating'] = pd.to_numeric(df['avg_rating'], errors='coerce')

# Calculate the average of 'avg_rating' by 'city'
avg_ratings_by_city = df.groupby('city')['avg_rating'].mean().reset_index()

# Select the top 20 cities with the highest average 'avg_rating'
top_20_cities = avg_ratings_by_city.nlargest(20, 'avg_rating')

# Set Seaborn theme
sns.set_theme(style="whitegrid")

# Bar plot for the top 20 cities with the highest average 'avg_rating'
plt.figure(figsize=(10, 5))
sns.barplot(x='city', y='avg_rating', data=top_20_cities, ci=None, color='green')
plt.title('Top 20 Cities with the Highest Average Ratings (Bar Plot)')
plt.xticks(rotation=45)
plt.show()
```

### Yelp_hotel_reviews 2/4

```python
import pandas as pd
```

```python
# import data
#yelp_hotel_reviews_viz = pd.read_csv("Datos\\Yelp\\Nuevos\\yelp_hotel_reviews.csv")
```

```python
# import data
yelp_hotel_reviews_viz = pd.read_csv("Datos\\Yelp\\drive\\yelp_hotel_reviews.csv")
```

```python
# Creando df
df_yelp_hotel_reviews_viz = pd.DataFrame(yelp_hotel_reviews_viz)
```

```python
df_yelp_hotel_reviews_viz.head()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df_yelp_hotel_reviews_viz is your DataFrame

# Convert 'date' column to datetime
df_yelp_hotel_reviews_viz['date'] = pd.to_datetime(df_yelp_hotel_reviews_viz['date'])

# Extract the year from the 'date' column
df_yelp_hotel_reviews_viz['year'] = df_yelp_hotel_reviews_viz['date'].dt.year

# Set the figure size
plt.figure(figsize=(12, 6))

# Create a bar plot of the number of reviews over time by year
sns.countplot(x='year', data=df_yelp_hotel_reviews_viz, color='orange')

# Add titles and labels
plt.title('Number of Reviews Over Time (By Year)')
plt.xlabel('Year')
plt.ylabel('Number of Reviews')

# Show the plot
plt.show()
```
