#### HOTELS EXTRACTION PROCESS

This notebook was used to execute manually the scripts to scrap all the hotels from the ten most populated cities of each state.

From the [scrapping.py](scrapping.py) file we import the function scrape_cities_hotels, wich takes a list of cities and returns a dataframe with the following hotel data:
- state
- city
- name
- avg_score
- price
- reviews_url

```python
import sys
import scrapping
import pandas as pd

sys.path.append('../files/lib')
# import personalised_graphics 
import selenium # personal library

usa_cities = pd.read_csv("../files/data/usa_cities.csv")
usa_states = pd.read_csv("../files/data/usa_states.csv")

# This functions takes a state_id and executes the scrapping process
def scrape(state_id: list): 
    state = state=usa_states["state"][usa_states["state_id"] == state_id].iloc[0] # find state name
    cities = usa_cities[(usa_cities["state_id"] == state_id)].sort_values(by="population", ascending=False).head(10) # filtrate top 10 populated cities
    hotels = scrapping.scrape_cities_hotels(cities["city"].to_list(), state, interfase=True) # scrapear hoteles
    #personalised_graphics.missing_values_heatmap(hotels, f"{state_id} scrape result") # explore data
    return hotels
```
