### visualization

```python
import numpy as np
import pandas as pd
import plotly.express as px

client_reviews = pd.read_csv("../../files/data/usa_clients_reviews.csv", index_col=0)
client_reviews
```

```python


fig = px.scatter(client_reviews, x="date", y="sentiment",
                 color='sentiment',
                 #labels={"sentiment": "Sentimiento de la Reseña"},
                 height=900)

# Mostrar la figura
fig.show()
```

```python


# Crear el Violin Plot con Plotly Express
fig = px.violin(client_reviews, x='company', y='sentiment', #color='', #points="all",
                title="Violin Plot de dos grupos",
                labels={"Value": "Valor", "Category": "Grupo"},
                violinmode="overlay",  # "overlay" para superponer
                box=False,  # No mostrar la caja
                height=900,  # Altura del gráfico
                width=1600
                )  # Ancho del gráfico

# Mostrar la figura
fig.show()
```
