```python
import pandas as pd

california_hotels = pd.read_csv("../../files/data/booking/california_hotels.csv", index_col=0)
california_hotels.drop_duplicates(subset=['city', 'name'], inplace=True)
california_hotels.dropna(subset=["scores"], inplace=True)
```

```python
import pandas as pd
import json

# Reemplazar comillas simples por comillas dobles
california_hotels['scores'] = california_hotels['scores'].str.replace("'", "\"")

# Convertir las cadenas JSON a listas de diccionarios
california_hotels['scores'] = california_hotels['scores'].apply(lambda x: json.loads(x) if pd.notna(x) else [])

# Crear columnas para cada clave del diccionario
df_scores_expanded = pd.json_normalize(california_hotels['scores'])

# Concatenar las columnas de puntajes expandidas al DataFrame original
california_hotels = pd.concat([california_hotels, df_scores_expanded], axis=1)

# Convertir las últimas 8 columnas a cadenas
california_hotels[california_hotels.columns[-8:]] = california_hotels[california_hotels.columns[-8:]].astype(str)

# Eliminar las llaves y comillas simples de las últimas 8 columnas
for col in california_hotels.columns[-8:]:
    california_hotels[col] = california_hotels[col].apply(lambda x: x.replace("{", "").replace("}", "").replace("'", "") if pd.notna(x) else None)


import pandas as pd

# Definimos una función para obtener el título de una celda
def get_title(cell_value):
    if pd.notna(cell_value):
        return cell_value.split(':')[0]
    else:
        return None

# Iteramos sobre las últimas 8 columnas y asignamos nuevos nombres
for col in california_hotels.columns[-8:]:
    # Usamos apply para obtener el título de cada celda
    new_titles = california_hotels[col].apply(get_title)
    
    # Asignamos los nuevos nombres a las columnas
    california_hotels = california_hotels.rename(columns=dict(zip([col]*len(new_titles), new_titles)))

# Ahora, df_combined tiene las columnas renombradas según los valores presentes en ellas

# Obtén el nombre de la anteúltima columna
anteultimate_col_index = -2
anteultimate_col_name = california_hotels.columns[anteultimate_col_index]

# Copia el nombre de las columnas
new_columns = list(california_hotels.columns)

# Cambia el nombre de la anteúltima columna a 'WiFi Gratis'
new_columns[anteultimate_col_index] = 'WiFi Gratis'

# Actualiza los nombres de las columnas en el DataFrame
california_hotels.columns = new_columns

import pandas as pd

# Selecciona solo las últimas 8 columnas
last_8_columns = california_hotels.iloc[:, -8:]

# Aplica una función para extraer y convertir los valores numéricos después de los dos puntos (:) en cada celda
numeric_values_only = last_8_columns.applymap(lambda x: pd.to_numeric(x.split(':')[-1], errors='coerce') if isinstance(x, str) and ':' in x else x)

# Actualiza las últimas 8 columnas en el DataFrame original con los valores numéricos
california_hotels.iloc[:, -8:] = numeric_values_only

california_hotels.replace("nan", None, inplace=True)

california_hotels
```

```python
california_hotels.dropna(subset=["city", "scores", "Personal"], inplace=True)
california_hotels.drop(columns=["None", "scores"], inplace=True)

california_hotels
```

```python
import plotly.express as px
import plotly.graph_objects as go

california_hotels = pd.read_csv("../../files/data/booking/california_hotels.csv", index_col=0)
california_hotels.drop_duplicates(subset=['city', 'name'], inplace=True)
california_hotels

default_center = {"lat": 36.7783, "lon": -119.4179}  # Example center for California
california_bbox = {"lon_min": -124.55, "lon_max": -114.13, "lat_min": 32.53, "lat_max": 42.0,}

cities_map = px.scatter_geo(
    california_hotels,
    lat="latitude",
    lon="longitude",
    color_discrete_sequence=["blue"],
    title="Geographic distribution:",
    width=1000,
    height=900,
    scope="usa",
    center=default_center,
    )

cities_map.update_traces(marker=dict(line=dict(width=0)), opacity=0.7)
cities_map.update_layout(title_x=0.7)
cities_map.update_geos(
    center_lon=default_center["lon"],
    center_lat=default_center["lat"],
    lonaxis_range=[california_bbox["lon_min"], california_bbox["lon_max"]],
    lataxis_range=[california_bbox["lat_min"], california_bbox["lat_max"]],
)


cities_map.show()
```
