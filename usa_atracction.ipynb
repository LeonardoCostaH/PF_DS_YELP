## Exploracion de categorias y atracciones turisticas

```python
df=df_usa_attractions.copy()
```

```python
# value_counts para contar la frecuencia de cada atracción turística
top_attractions = df['attraction'].value_counts().nlargest(10)

# Visualización de las atracciones turísticas más relevantes
plt.figure(figsize=(10, 6))
top_attractions.plot(kind='bar', color='skyblue')
plt.title('Top 10 Atracciones Turísticas más Relevantes')
plt.xlabel('Atracción Turística')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# value_counts para contar la frecuencia de cada categoría
top_categories = df['categories'].value_counts().nlargest(10)

# Visualización de las categorías más relevantes
plt.figure(figsize=(10, 6))
top_categories.plot(kind='bar', color='salmon')
plt.title('Top 10 Categorías más Relevantes')
plt.xlabel('Categoría')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# value_counts para contar la frecuencia de atracciones turísticas por estado
attractions_by_state = df['state'].value_counts()

# Visualización de atracciones turísticas por estado
plt.figure(figsize=(12, 6))
attractions_by_state.plot(kind='bar', color='green')
plt.title('Atracciones Turísticas por Estado')
plt.xlabel('Estado')
plt.ylabel('Número de Atracciones Turísticas')
plt.show()
```

```python
# value_counts para contar la frecuencia de atracciones turísticas por estado
attractions_by_state = df['state'].value_counts()

# Visualización de atracciones turísticas por estado
plt.figure(figsize=(12, 6))
attractions_by_state.plot(kind='bar', color='green')
plt.title('Atracciones Turísticas por Estado')
plt.xlabel('Estado')
plt.ylabel('Número de Atracciones Turísticas')
plt.show()
```

```python
top_states = df['state'].value_counts().nlargest(10)

# Visualización de los estados con más atracciones turísticas
plt.figure(figsize=(10, 6))
top_states.plot(kind='bar', color='orange')
plt.title('Top 10 Estados con más Atracciones Turísticas')
plt.xlabel('Estado')
plt.ylabel('Número de Atracciones Turísticas')
plt.show()
```

## Analisis segun las reviews

```python
df['n_reviews'].describe()
```

```python
# Agrupa por categoría y calcula la suma de revisiones para cada categoría
top_categories_reviews = df.groupby('categories')['n_reviews'].sum().nlargest(10)

# Visualización del top ten de categorías con más revisiones
plt.figure(figsize=(12, 6))
top_categories_reviews.plot(kind='bar', color='salmon')
plt.title('Top 10 Categorías con más Revisiones')
plt.xlabel('Categoría')
plt.ylabel('Número Total de Reviews')
plt.show()
```
